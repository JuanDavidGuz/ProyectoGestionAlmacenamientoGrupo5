```python
import pandas as pd
import matplotlib.pyplot as plt
```

Se trabaja con un conjunto de datos proporcionados por el Sistema Nacional de Información de la Educación Superior (SNIES) el cual nos brinda información referente a los estudiantes graduados en instituciones de educación superior en Colombia en el año 2023.

```python
#Carga de Datos
xlsx = "Estudiantes.xlsx"
df = pd.read_excel(xlsx, sheet_name=1, skiprows=5)
df = df.dropna(thresh=5)
df
```

#Descriptivas


```python
print(f"El dataset tiene {df.shape[0]} filas y {df.shape[1]} columnas")
```

```python
print(pd.DataFrame(df.columns, columns=['Column Names']))
```

```python
print("Tipos de datos en el dataset:")
print(df.dtypes)
```

```python
universidades_total = df["INSTITUCIÓN DE EDUCACIÓN SUPERIOR (IES)"].nunique()
departamentos_total = df["DEPARTAMENTO DE DOMICILIO DE LA IES"].nunique()
sector = df["SECTOR IES"].value_counts()
programas_total = df["DESC CINE CAMPO DETALLADO"].nunique()
programas = df["DESC CINE CAMPO DETALLADO"].value_counts()

print(f" Total de universidades registradas: {universidades_total}")
print(f" Total de departamentos registrados en el dataset: {departamentos_total}\n")

print("Distribución de Universidades por Sector")
print(sector, "\n")

print(f"Total de programas académicos registrados: {programas_total}")
print("\n Top de programas académicos registrados:")
print(programas.head(10))  # Mostrar solo los 10 programas más registrados
```

```python
df["NIVEL ACADÉMICO"].value_counts().plot(kind="bar", color='lightgreen')
plt.title("Distribución de Universidades por Nivel Académico")
plt.ylabel("Cantidad de Universidades")
plt.xlabel("Nivel Académico")
plt.show()
```

```python
universidades_por_region = df["DEPARTAMENTO DE DOMICILIO DE LA IES"].value_counts()
print("Cantidad de universidades por región:")
print(universidades_por_region)
universidades_por_region.plot(kind="bar", figsize=(10, 6), color='lightgreen')
plt.title("Distribución de Universidades por Región")
plt.ylabel("Cantidad de Universidades")
plt.xlabel("Región")
plt.show()
```

```python
graduados_por_region =df.groupby("DEPARTAMENTO DE DOMICILIO DE LA IES")["GRADUADOS"].sum()

comparacion =pd.DataFrame({"Graduados": graduados_por_region})

print("Cantidad de graduados por región:")

print(comparacion)
```

```python
graduados_por_region =df.groupby("DEPARTAMENTO DE DOMICILIO DE LA IES")["GRADUADOS"].sum().plot(kind="bar", figsize=(10, 6), color='lightblue')
plt.title("Cantidad de graduados por región")
plt.ylabel("Cantidad de graduados")
plt.xlabel("Región")
plt.show()
```
