```python
menos_graduados = df.groupby("DEPARTAMENTO DE DOMICILIO DE LA IES")["GRADUADOS"].sum().nsmallest(6)

print("Los 6 departamentos con menor número de graduados son:")
menos_graduados
```

```python

menos_graduados = df.groupby("DEPARTAMENTO DE DOMICILIO DE LA IES")["GRADUADOS"].sum().nsmallest(6)
total_graduados = df["GRADUADOS"].sum()


plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['axes.grid'] = False
fig, ax = plt.subplots(figsize=(12, 6))


colors = plt.cm.tab20(np.linspace(0, 1, len(menos_graduados)))


bars = ax.barh(
    menos_graduados.index,
    menos_graduados.values,
    color=colors,
    edgecolor='white',
    linewidth=1.5,
    height=0.6
)


for i, (depto, count) in enumerate(menos_graduados.items()):
    ax.text(
        count + (max(menos_graduados.values)*0.03),
        i,
        f'{count:,} ({count/total_graduados*100:.1f}%)',
        va='center',
        fontsize=10,
        weight='bold',
        color='black'
    )


ax.set_xlim(0, max(menos_graduados.values)*1.2)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(True)


plt.title(
    '6 Departamentos con Menor Número de Graduados\n',
    fontsize=14,
    fontweight='bold',
    pad=20,
    color='black'
)
plt.xlabel('Número de Graduados', fontsize=11)
plt.ylabel('')
plt.tight_layout()
plt.show()
plt.show()
```
